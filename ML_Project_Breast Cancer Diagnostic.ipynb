```python
#min max normalizasyonunun uygulanması
#perimeter_mean,area_mean özellikleri 0 ile 1 arasında normalize edilmesi
from sklearn.preprocessing import minmax_scale
import pandas as pd
hw_scaled = minmax_scale(breast_cancer[['perimeter_mean','area_mean']], feature_range=(0,1))
 
breast_cancer['perimeter_mean']=tuple(hw_scaled[:,0])
breast_cancer['area_mean']=tuple(hw_scaled[:,1])

```

```python
#Özellik Seçim Yöntemi: SelectKBest
#Score Fonksiyonu: chi2
#SelectKBest supervised modelleri için tercih edilir. 
#chi2 sınıflandırma(classification) problemleri için kullanılır. 
from sklearn.feature_selection import SelectKBest,  chi2
chi2_selector = SelectKBest(chi2, k='all')
chi2_selector.fit(breast_cancer.drop(columns=['diagnosis']), breast_cancer['diagnosis'])
p_values = pd.Series(chi2_selector.pvalues_)
p_values.index=breast_cancer.drop(columns=['diagnosis']).columns
p_values.sort_values(ascending=True,inplace=True)
print(p_values)
p_values.plot.bar()
plt.yticks([i/20 for i in range(20)])
plt.show()
uninformative = p_values[p_values>0.01]#p value değeri 0.01'den büyük olan değerler veri setimizdeki önemsiz özelliklerdir 
print('Uninformative columns:\n',uninformative)#önemsiz özellikler
```
